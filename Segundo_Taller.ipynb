```python
dfy=df[["wage"]]
dfy
```

```python
dfx=df[["exper","educ"]]
dfx
```

```python
arr = dfy.to_numpy()
print(arr)
```

```python
arr = dfx.to_numpy()
print(arr)
```

```python
arr2 = dfy.to_numpy()
print(arr2)
```

```python
X=np.c_[np.ones(len(dfx)),dfx]
X
```

```python
betas=np.linalg.inv(X.T.dot(X)).dot(X.T.dot(dfy))
betas.squeeze
betas

```

```python
y_est=(X@betas)
y_est
```

```python
u_est=arr2-y_est
u_est
```

```python
Var=(u_est.T.dot(u_est))/523
Var
```

```python
ee=np.diag(np.sqrt(Var*(np.linalg.inv(X.T.dot(X)))))
ee 
```

```python
est_t=betas/ee
est_t
```
